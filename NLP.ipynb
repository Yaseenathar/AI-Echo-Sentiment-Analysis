```python
# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re, string
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Convert ratings (15) to sentiment
def rating_to_sentiment(r):
    if r <= 2:
        return "Negative"
    elif r == 3:
        return "Neutral"
    else:
        return "Positive"

df["sentiment"] = df["rating"].apply(rating_to_sentiment)
print("Class distribution:\n", df["sentiment"].value_counts())

# Download NLTK resources
nltk.download("stopwords")
nltk.download("punkt")
nltk.download("wordnet")

stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+", "", text)   # remove URLs
    text = text.translate(str.maketrans("", "", string.punctuation))  # remove punctuation
    tokens = nltk.word_tokenize(text)
    tokens = [lemmatizer.lemmatize(w) for w in tokens if w.isalpha() and w not in stop_words]
    return " ".join(tokens)

df["clean_review"] = df["review"].apply(clean_text)

# --- Sentiment distribution ---
sns.countplot(x="sentiment", data=df)
plt.title("Sentiment Distribution")
plt.show()

# --- Word cloud for positive reviews ---
text_pos = " ".join(df[df["sentiment"]=="Positive"]["clean_review"])
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text_pos)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Positive Review WordCloud")
plt.show()

# --- Features ---
X = df["clean_review"]
y = df["sentiment"]

tfidf = TfidfVectorizer(max_features=5000)
X_tfidf = tfidf.fit_transform(X)

# --- Logistic Regression with class weights ---
model = LogisticRegression(max_iter=500, class_weight="balanced")

# Instead of single train/test split, use stratified cross-validation
y_pred = cross_val_predict(model, X_tfidf, y, cv=5)

# --- Classification Report ---
print("\nClassification Report (5-fold CV):\n")
print(classification_report(y, y_pred))

# --- Confusion Matrix ---
cm = confusion_matrix(y, y_pred, labels=["Positive","Neutral","Negative"])
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Positive","Neutral","Negative"],
            yticklabels=["Positive","Neutral","Negative"])
plt.title("Confusion Matrix")
plt.show()

# --- Prediction function ---
def predict_sentiment(review_text):
    clean = clean_text(review_text)
    vec = tfidf.transform([clean])
    return model.fit(X_tfidf, y).predict(vec)[0]  # fit on full data for deployment

print("\nSample Predictions:")
print(predict_sentiment("This app is amazing, I use it every day!"))
print(predict_sentiment("It gives wrong answers and crashes often."))
print(predict_sentiment("It is okay, nothing special but not bad either."))
```

```python
import torch
from torch import nn, optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import LabelEncoder
from keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences


# Encode labels
le = LabelEncoder()
y_dl = le.fit_transform(df["sentiment"])

# Tokenize
tokenizer = Tokenizer(num_words=10000, oov_token="<OOV>")
tokenizer.fit_on_texts(df["clean_review"])
X_seq = tokenizer.texts_to_sequences(df["clean_review"])

# Pad sequences
X_pad = pad_sequences(X_seq, maxlen=100, padding="post", truncating="post")

# Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_pad, y_dl, test_size=0.2, random_state=42, stratify=y_dl)

# Convert to tensors
X_train_t = torch.tensor(X_train, dtype=torch.long)
y_train_t = torch.tensor(y_train, dtype=torch.long)
X_test_t = torch.tensor(X_test, dtype=torch.long)
y_test_t = torch.tensor(y_test, dtype=torch.long)

train_ds = TensorDataset(X_train_t, y_train_t)
train_dl = DataLoader(train_ds, batch_size=32, shuffle=True)

# LSTM Model
class LSTMModel(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, output_dim, num_layers=2, dropout=0.5):
        super(LSTMModel, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
        self.lstm = nn.LSTM(embed_dim, hidden_dim, num_layers=num_layers,
                            dropout=dropout, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(hidden_dim*2, output_dim)  # *2 for bidirectional
        self.dropout = nn.Dropout(0.5)
    
    def forward(self, x):
        x = self.embedding(x)
        _, (h, _) = self.lstm(x)
        h = torch.cat((h[-2], h[-1]), dim=1)  # concat forward + backward
        h = self.dropout(h)
        return self.fc(h)

# Adjust vocab size properly
vocab_size = min(len(tokenizer.word_index)+1, 10000)
output_dim = len(set(y_dl))  # Automatically set based on labels

model = LSTMModel(vocab_size, embed_dim=128, hidden_dim=256, output_dim=output_dim)

loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
for epoch in range(20):
    model.train()
    total_loss = 0
    for xb, yb in train_dl:
        optimizer.zero_grad()
        preds = model(xb)
        loss = loss_fn(preds, yb)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1} Loss: {total_loss/len(train_dl):.4f}")
```
